```python
import time
import pygame
import random
import tensorflow as tf
import numpy as np
from collections import deque
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
```

```python
BLOCK = 10
SPEED = 60
MAX_SPEED = 20
RANDOM = 1

class PingPong:
    def __init__(self, w=600, h=400):
        pygame.init()
        self.w = w 
        self.h = h
        self.paddle_length = BLOCK*5
        self.ball_radius = 7
        self.hit_right_this_step = False
        self.font = pygame.font.SysFont('Arial', 30)
        self.display = pygame.display.set_mode((w, h))
        pygame.display.set_caption("Ping Pong RL Environment")
        self.clock = pygame.time.Clock()
        self.reset()

    
    def reset(self):
        self.score1 = 0
        self.score2 = 0
        self.ball_x = random.randint(self.w//3, 2*self.w//3)
        self.ball_y = random.randint(0, self.h)
        self.ball_vx = random.choice([-3,3])
        self.ball_vy = random.uniform(-2,2)
        self.Lpaddle_y = self.h/2 - self.paddle_length/2
        self.Rpaddle_y = self.h/2 - self.paddle_length/2

        return self.get_state()

    def reset_rally(self):
        self.ball_x = random.randint(self.w//3, 2*self.w//3)
        self.ball_y = random.randint(0, self.h)
        self.ball_vx = random.choice([-3,3])
        self.ball_vy = random.uniform(-2,2)
        self.Lpaddle_y = self.h/2 - self.paddle_length/2
        self.Rpaddle_y = self.h/2 - self.paddle_length/2

        return self.get_state()
    
    def render(self):
        self.display.fill((0,0,0))
    
        # Left paddle
        pygame.draw.rect(
            self.display,
            (255,255,255),
            pygame.Rect(10, self.Lpaddle_y , BLOCK, self.paddle_length)
        )
    
        # Right paddle
        pygame.draw.rect(
            self.display,
            (255,255,255),
            pygame.Rect(self.w-20, self.Rpaddle_y , BLOCK, self.paddle_length)
        )
    
        # Ball
        pygame.draw.circle(
            self.display,
            (255,255,255),
            (int(self.ball_x), int(self.ball_y)),
            self.ball_radius
        )

        # middle line
        pygame.draw.rect(
            self.display,
            (255,255,255),
            pygame.Rect(self.w/2, 0 , 1 , self.h)
        )

        # score
        score_txt = f"{self.score1}      {self.score2}"
        text_surface = self.font.render(score_txt, True, (255, 255, 255))
        self.display.blit(text_surface, (self.w/2 - 40, 20))

        self.clock.tick(SPEED)
    
        pygame.display.flip()

    def get_state(self):
        state = np.array([
            self.ball_x/self.w,
            self.ball_y/self.h,
            self.ball_vx/MAX_SPEED,
            self.ball_vy/MAX_SPEED,
            self.Rpaddle_y/(self.h-self.paddle_length),
            self.Lpaddle_y/(self.h-self.paddle_length)
            ],dtype=np.float32)
        
        return state.reshape(1,-1)

    def step(self, action):
        '''
            [1,0,0] --> move up
            [0,1,0] --> no move
            [0,0,1] --> move down
        '''
        self.hit_right_this_step = False
        # move left paddle
        self.move_opponent()

        # take actioin on right paddle
        if action[0]==1:
            self.Rpaddle_y = max(0,self.Rpaddle_y - BLOCK)
        if action[2]==1:
            self.Rpaddle_y = min(self.h - self.paddle_length,self.Rpaddle_y + BLOCK)

        
        # move the ball one frame
        self.ball_x += self.ball_vx
        self.ball_y += self.ball_vy

        self.collision()
        reward, done = self.get_info()
        next_state = self.get_state()

        return next_state, reward, done

    def collision(self):
        # top/bottom
        if self.ball_y - self.ball_radius <= 0 or self.ball_y + self.ball_radius >= self.h:
            self.ball_vy *= -1
    
        # LEFT PADDLE
        if self.ball_x - self.ball_radius <= 10 + BLOCK:
            if self.Lpaddle_y <= self.ball_y - self.ball_radius//2 <= self.Lpaddle_y + self.paddle_length:
                self.ball_vx = abs(self.ball_vx) * 1.1
                offset = (self.ball_y - (self.Lpaddle_y + self.paddle_length/2)) / (self.paddle_length/2)
                self.ball_vy += offset * 2
    
        # RIGHT PADDLE  ← here we set the flag
        if self.ball_x + self.ball_radius >= self.w - 20:
            if self.Rpaddle_y <= self.ball_y  <= self.Rpaddle_y + self.paddle_length:
                self.hit_right_this_step = True
                self.ball_vx = -abs(self.ball_vx) * 1.1
                offset = (self.ball_y - (self.Rpaddle_y + self.paddle_length/2)) / (self.paddle_length/2)
                self.ball_vy += offset * 2
    
        self.ball_vx = max(-MAX_SPEED, min(MAX_SPEED, self.ball_vx))
        self.ball_vy = max(-MAX_SPEED, min(MAX_SPEED, self.ball_vy))


    def get_info(self):
        reward = 0.0
        done = False
    
        # left side miss (opponent scores)
        if self.ball_x < 20 and not (self.Lpaddle_y <= self.ball_y <= self.Lpaddle_y + self.paddle_length):
            done = True
            reward = 10
            self.score2 += 1
    
        # right side miss (agent loses point)
        elif self.ball_x > self.w - 30 and not (self.Rpaddle_y <= self.ball_y <= self.Rpaddle_y + self.paddle_length):
            done = True
            reward = -10
            self.score1 += 1
    
        # right paddle hit → give shaped reward
        elif self.hit_right_this_step:
            dist = abs(self.ball_y - (self.Rpaddle_y +self.paddle_length/2))
            reward = 0.5 + (dist/(self.paddle_length/2))*1.5
    
        # small living reward (helps long rallies)
        else:
            paddle_center = self.Rpaddle_y + self.paddle_length / 2
            dist = abs(paddle_center - self.ball_y)
            reward = 0.03 * (1 - min(dist / 80, 1.0)) 
    
        return reward, done

    def move_opponent(self):
        if random.random() < RANDOM:
            return
    
        paddle_center = self.Lpaddle_y + self.paddle_length / 2
        diff = paddle_center - self.ball_y
        deadzone = 10
    
        if diff > deadzone:
            self.Lpaddle_y = max(0, self.Lpaddle_y - 0.7 * BLOCK)
        elif diff < -deadzone:
            self.Lpaddle_y = min(self.h - self.paddle_length, self.Lpaddle_y + 0.7 * BLOCK)
            
    def move_agent(action):
        if action[0]==1:
            self.Rpaddle_y = max(0,self.Rpaddle_y - BLOCK)
        if action[2]==1:
            self.Rpaddle_y = min(self.h - self.paddle_length,self.Rpaddle_y + BLOCK)

    def handle_events(self):
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                
    def isFinished(self):
        if self.score1==5 or self.score2==5:
            return True
        return False

    def move_user(self):
        keys = pygame.key.get_pressed()
        
        if keys[pygame.K_UP]: # Move Up
            self.Lpaddle_y = max(0,self.Lpaddle_y - BLOCK)
        elif keys[pygame.K_DOWN]: # Move Down
            self.Lpaddle_y = min(self.h - self.paddle_length, self.Lpaddle_y + BLOCK)           
```

```python
# Deep Q-learning environment

model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(6,)),
    tf.keras.layers.Dense(units = 64, activation='relu'),
    tf.keras.layers.Dense(units = 64, activation='relu'),
    tf.keras.layers.Dense(units = 3, activation='linear')
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
    loss='mse'
)

target_model = tf.keras.models.clone_model(model)
target_model.set_weights(model.get_weights())

replay_buffer = deque(maxlen=500_000)
```

```python
TOTAL_STEPS_BUDGET = 2_000_000
REPLAY_MIN_SIZE    = 10_000
BATCH_SIZE         = 32
TRAIN_INTERVAL     = 8
TARGET_UPDATE_FREQ = 7_000
MAX_FRAMES_EP      = 3500
EPS_DECAY_STEPS    = 700_000
RANDOM_DECAY_STEPS = 700_000
EPS_MIN            = 0.01

epsilon = 1.0
gamma   = 0.98
total_steps = 0
episode = 0
start_time = time.time()

win_history = []
lose_history = []
avg_reward_history = []
episode_lengths = []

game = PingPong()

while total_steps < TOTAL_STEPS_BUDGET:
    episode += 1
    state = game.reset()
    done = False
    frames = 0
    episode_reward = 0.0
    hits_this_episode = 0
        
    while not done:

        # Epsilon-greedy action
        if random.random() < epsilon:
            action_idx = random.randrange(3)
        else:
            q = model(state, training=False)[0]
            action_idx = np.argmax(q)

        action_onehot = np.eye(3)[action_idx]
        next_state, reward, done = game.step(action_onehot)
        
        if episode%100 == 0:
            game.render()
            
        replay_buffer.append((state, action_idx, reward, next_state, done))
        state = next_state

        episode_reward += reward
        if game.hit_right_this_step == True:
            hits_this_episode += 1
        
        if len(replay_buffer) >= REPLAY_MIN_SIZE and total_steps % TRAIN_INTERVAL == 0:
            batch = random.sample(replay_buffer, 32)
            state_, action_idx_, reward_, next_state_, done_ = zip(*batch)

            action_ = np.array(action_idx_)
            reward_ = np.array(reward_)
            state_ = np.vstack(state_)
            next_state_ = np.vstack(next_state_)
            done_   = np.array(done_, dtype=int)

            q_values = model(state_, training=False).numpy()
            q_next   = target_model(next_state_, training=False).numpy()
            
            targets = q_values.copy()
            
            max_q_next = np.max(q_next, axis=1)
            
            targets[np.arange(len(batch)), action_] = reward_ + (1 - done_) * gamma * max_q_next
            
            model.fit(state_, targets, epochs=1, verbose=0)

        total_steps += 1
        frames += 1
        
        RANDOM = max(0.1, 1.0 - total_steps / RANDOM_DECAY_STEPS)
        
        # Decay epsilon linearly
        epsilon = max(EPS_MIN, 1.0 - total_steps / EPS_DECAY_STEPS)
        
        game.handle_events()

        if total_steps % TARGET_UPDATE_FREQ == 0:
            target_model.set_weights(model.get_weights())
    
        # Progress logging
        if total_steps % 10000 == 0:
            elapsed = time.time() - start_time
            print(f"step {total_steps:7,d} | ep {episode:5,d} | "
                  f"eps {epsilon:.3f} | "
                  f"time {elapsed:.1f}s")
            start_time = time.time()
        
        if frames >= MAX_FRAMES_EP:
            done = True
        frames += 1
        
    # Episode summary
    win_history.append(game.score2)
    lose_history.append(game.score1)
    avg_reward_history.append(episode_reward / frames if frames > 0 else 0)
    episode_lengths.append(frames)

    if episode % 100 == 0:
        recent = win_history[-100:]
        recent_loss = lose_history[-100:]
        print(f"Ep {episode:5,d} | "
              f"Avg win rate : {np.mean(recent)*100:.2f}  | "
              f"Avg lose rate: {np.mean(recent_loss)*100:.2f}  | "
              f"Avg reward/step: {np.mean(avg_reward_history[-100:]):.4f}  | "
              f"Avg length: {np.mean(episode_lengths[-100:]):.0f}")
```

```python
#plot figures
group_size = 100

def group_avg(arr, k):
    return [
        np.mean(arr[i:i+k])
        for i in range(0, len(arr) - k + 1, k)
    ]

# ---- Group all ----
win_grouped = group_avg(win_history, group_size)
reward_grouped = group_avg(avg_reward_history, group_size)
len_grouped = group_avg(episode_lengths, group_size)

# ---- Create x-axes ----
groups = np.arange(1, len(win_grouped) + 1) * group_size

# ---- Plot ----
plt.figure(figsize=(15, 5))

# 1. Win history
plt.subplot(1, 3, 1)
plt.plot(groups, win_grouped)
plt.title("Win History (per 100 episodes)")
plt.xlabel("Episodes")
plt.ylabel("Avg Score")

# 2. Avg reward history
plt.subplot(1, 3, 2)
plt.plot(groups, reward_grouped)
plt.title("Average Reward (per 100 episodes)")
plt.xlabel("Episodes")
plt.ylabel("Reward")

# 3. Episode lengths
plt.subplot(1, 3, 3)
plt.plot(groups, len_grouped)
plt.title("Episode Lengths (per 100 episodes)")
plt.xlabel("Episodes")
plt.ylabel("Frames")

plt.tight_layout()
plt.savefig('plots.png', dpi=300)
plt.show()
```

```python
#playing
game = PingPong()
state = game.reset()

running = True

while running:
    game.handle_events()

    # ---- HUMAN CONTROL (LEFT PADDLE) ----
    # game.move_user()
    game.move_opponent()

    # ---- AGENT CONTROL (RIGHT PADDLE) ----
    q_values = target_model(state, training=False)[0].numpy()
    action_idx = np.argmax(q_values)

    action = np.eye(3)[action_idx]

    # ---- STEP ENV ----
    next_state, reward, done = game.step(action)
    state = next_state

    # ---- RENDER ----
    game.render()

    # ---- RESET ROUND IF POINT SCORED ----
    if done:
        wait_time = 1000  # ms
        start_wait = pygame.time.get_ticks()
    
        while pygame.time.get_ticks() - start_wait < wait_time:
            game.handle_events()   # still allow window close
            game.render()          # keep showing current frame

        state = game.reset_rally()

    # ---- END GAME IF SOMEONE WINS ----
    if game.isFinished():
        print(f"Final Score -> You: {game.score1} | Agent: {game.score2}")
        running = False

pygame.quit()
```
